```python
data=data.drop('Unnamed: 0', axis =1)
data
```

```python
# get numerical data
numerical_columns = data.select_dtypes(exclude=['object']).columns.tolist()
numerical_columns
```

```python
data_numerical = data[numerical_columns]
data_numerical
```

```python
#Use the describe() method on the numeric data in the DataFrame.
data_numerical.describe()
```

```python
summary = data_numerical.describe().T
summary
```

```python
# Print the median
data_numerical.median(axis=None, skipna=None, level=None, numeric_only=None)
```
