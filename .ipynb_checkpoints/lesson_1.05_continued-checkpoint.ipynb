```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
data = pd.read_csv('merged_clean_ver1.csv')
```

```python
def clean(x):
    if x in ['M', 'MALE']:
        return 'Male'
    elif x.startswith('F'):
        return 'Female'
    else:
        return 'U'
```

```python
data['gender'] = list(map(clean, data['gender'])) 
```

```python
data['gender'].unique()
```

```python
data.drop('Unnamed: 0', axis=1, inplace =  True)
```

# Correlation

-  In the broadest sense correlation is any statistical association, though it actually refers to the degree to which a pair of variables are linearly related.



# Correlation does not imply causation!!!
- What’s the difference?
    - Correlation describes an association between variables,i.e, when one variable changes, so does the other. 

    - Causation means that changes in one variable brings about changes in the other; there is a cause-and-effect relationship between variables. 

    - A correlation doesn’t imply causation, but causation always implies correlation.

# Categorical variables - Chi square test for independence

```python
help(pd.crosstab)
```

```python
crosstab = pd.crosstab(data['state'], data['gender'])  # note we need to clean column state before we use this 
# crosstab
```

```python
crosstab
```

```python
help( stats.chi2_contingency)
```

```python
[statistic, p_value, dof, arr] = stats.chi2_contingency(crosstab)
print(p_value)

# first value is the chi-sq test static, the second value is the p-value, 
# and the third number is the degrees of freedom
# An array is also returned which contains the expected cell counts
```

```python
[statistic, p_value, dof, arr] = stats.chi2_contingency(crosstab)
print(p_value)

# first value is the chi-sq test static, the second value is the p-value, 
# and the third number is the degrees of freedom
# An array is also returned which contains the expected cell counts
```
